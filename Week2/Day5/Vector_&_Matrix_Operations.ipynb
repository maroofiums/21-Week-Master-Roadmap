## Initalization

```python
import math
import numpy as np
```

```python
v = [1, 2, 3]
w = [4, 5, 6]
c = 2
```

# Vector Operations

## Vector Addition

```python
def vector_add(v,w):
  if len(v) != len(w):
    raise ValueError("Vector must be same length")
  return [v[i] + w[i] for i in range(len(v))]

vector_add(v,w)
```

## Scalar Multiplication

```python
def scalar_multiply(c, v):
    return [c * x for x in v]
scalar_multiply(c,v)
```

## Dot Product

```python
def dot_product(v, w):
    if len(v) != len(w):
        raise ValueError("Vectors must be same length")
    return sum(v[i] * w[i] for i in range(len(v)))

dot_product(v,w)
```

## Vector Magnitude

```python
def magnitude(v):
    return math.sqrt(sum(x**2 for x in v))
magnitude(v)
```

# Matrix Operations

```python
A = [
    [1, 2],
    [3, 4]
]
B = [
    [5, 6],
    [7, 8]
]
```

## Matrix Addition

```python
def matrix_add(A, B):
  if len(A) != len(B) or len(A[0]) != len(B[0]):
    raise ValueError("Matrices must have same dimensions")

  result = []
  for i in range(len(A)):
    row = []
    for j in range(len(A[0])):
      row.append(A[i][j] + B[i][j])
    result.append(row)

  return result

matrix_add(A,B)
```

## Matrix Transpose

```python
def transpose(A):
  rows = len(A)
  cols = len(A[0])

  result = []
  for j in range(cols):
    row = []
    for i in range(rows):
      row.append(A[i][j])
    result.append(row)

  return result

transpose(A)
```

## Matrix Multiplication

```python
def matrix_multiply(A, B):
  if len(A[0]) != len(B):
    raise ValueError("Invalid dimensions for multiplication")

  result = []
  for i in range(len(A)):
    row = []
    for j in range(len(B[0])):
      value = 0
      for k in range(len(B)):
        value += A[i][k] * B[k][j]
      row.append(value)
    result.append(row)

  return result

matrix_multiply(A,B)
```
